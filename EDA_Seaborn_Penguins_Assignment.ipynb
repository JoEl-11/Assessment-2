# Exploratory Data Analysis (EDA) using Seaborn

**Dataset:** `penguins` dataset from Seaborn



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()

penguins.head()
```

## Dataset Information

```python
penguins.info()
penguins.describe(include='all')
```

## Species Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=penguins,x='species')
plt.title('Species Distribution')
plt.show()
```

### Insight
- Adelie penguins are the most common species in the dataset.
- Gentoo and Chinstrap occur in smaller numbers.
- The dataset is slightly imbalanced across species.

## Flipper Length Distribution

```python
plt.figure(figsize=(7,4))
sns.histplot(data=penguins,x='flipper_length_mm',bins=20,kde=True)
plt.title('Distribution of Flipper Length')
plt.show()
```

### Insight
- Most penguins have flipper lengths between 180 mm and 220 mm.
- The distribution suggests multiple clusters corresponding to different species.

## Body Mass by Species

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=penguins,x='species',y='body_mass_g')
plt.title('Body Mass by Species')
plt.show()
```

### Insight
- Gentoo penguins are generally the heaviest.
- Adelie penguins have the lowest median body mass.
- Boxplots also reveal the spread within each species.

## Bill Length vs Bill Depth

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species')
plt.title('Bill Length vs Bill Depth')
plt.show()
```

### Insight
- Species form distinct clusters based on bill measurements.
- Bill length and depth help distinguish penguin species.

## Correlation Heatmap

```python
plt.figure(figsize=(6,5))
corr = penguins[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Insight
- Flipper length and body mass show a strong positive correlation.
- Bill depth has a weaker relationship with the other measurements.

# Conclusion

The Penguin dataset reveals clear differences among species. Gentoo penguins are generally larger and heavier, while Adelie penguins are smaller. Flipper length is strongly associated with body mass, and bill measurements effectively separate species. These visualizations demonstrate how EDA helps identify distributions, relationships, and patterns before building machine learning models.
